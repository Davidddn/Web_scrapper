```python
TP=int(input("Please enter the true positive value in numbers"))
TN=int(input("Please enter the true negative value in numbers"))
FP=int(input("Please enter the false positive value in numbers"))
FN=int(input("Please enter the fales negative value in numbers"))
```

```python
if(TP+FP == 0):
 print("Values are not applicable")
else:
 Precision = TP/(TP+FP)
```

```python
if(TP+FN == 0):
 print("Values are not applicable")
else:
 Recall=TP/(TP+FN)
```

```python
if(TP+FN+FP+TN == 0):
 print("Values are not applicable")
else:
 Accuracy=(TP+TN)/(TP+FN+FP+TN)
```

```python
print(Precision)
```

```python
print(Recall)
```

```python
print(Accuracy)
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://pureportal.coventry.ac.uk/en/organisations/school-of-computing-electronics-and-maths/persons/"
page=requests.get(url)
print(page.text)
```

```python
soup=BeautifulSoup(page.text,"html.parser")
```

```python
first_staff=soup.find("a", class_="link person")
print(first_staff)
```

```python
print(first_staff.string)
```

```python
all_staff=soup.find_all("a", class_="link person")
for this_staff in all_staff:
 print(this_staff.string)
```

```python
for a in soup.find_all("a", href=True, class_="link person"):
  print(a['href']) 
```

```python
for this_staff in all_staff:
 print(this_staff.string)
 print(this_staff['href'])
 print("---------------------------------------") 
```
